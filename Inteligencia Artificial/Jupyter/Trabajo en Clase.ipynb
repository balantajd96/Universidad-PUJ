# Arboles de Decisión y Bosques Aleatorios

## Punto 1
### Describir qué es un árbol de decisión


Un árbol de decisión es un modelo empleado para la clasificación y la regresión. Este modelo consiste en la ramificación de preguntas secuenciales y el retorno de una ruta del árbol dada una respuesta.

## Punto 2
### Describir qué es un bosque aleatorio

Un bosque aleatorio es una colección de árboles de decisión en el que todos los resultados de los arboles se agregan en un unico resultado final.

 ### Explicar para qué se utilizan funciones como GINI o Ganancia de información en la construcción de un árbol de decisión.

Gini es una medida de la desigualdad que sirve para indicar la frecuencia con la que una variable elegida al azar se clasificará erróneamente. Debido a que cuando construimos el árbol de decisión, queremos saber cuál será la pregunta correcta para comenzar.
Inicialmente se divide el conjunto de datos, luego elegimos una instancia al azar y hayamos la probabilidad de se clasifique mal.

## Punto 6

### Explicar porqué es mejor un bosque aleatorio que un árbol de decisión.
El bosque aleatorio reducirá la parte de varianza del error en lugar de la parte de sesgo, por lo que en un determinado conjunto de datos de entrenamiento el árbol de decisión puede ser más preciso que un bosque aleatorio. Pero en un conjunto de datos de validación inesperado, el bosque aleatorio siempre gana en términos de precisión.
