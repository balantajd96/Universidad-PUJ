#### Objetivo
- Poder secuenciar las operaciones en las maquinas de tal manera que se minimice el tiempo Cmax, requerido para la finalización de todos los trabajos.

#### Variables del Problema
- J  conjunto de n trabajos.
- M conjunto de m máquinas.
- ni son operaciones  consecutivas donde i es cada trabajo.
- La  j-esima  operación del trabajo i, se denota por Oij.
- Cada operacion Oij puede ser procesada por las maquinas pertenecientes Mij ⊆ M de máquinas elegibles.
- Para cada operacion Oij, sea Pijk su tiempo de procesamiento en la máquina k ∈ Mij.

#### Restricciones
- Cada operación debe completarse sin interrupción en la misma máquina una vez inicia.
- Cada operación se asigna a una sóla máquina elegible.
- Las máquinas no pueden realizar más de una operación a la vez.
- Las operaciones que componen un trabajo deben realizarse estrictamente en el orden establecido al definir el trabajo.
- Todos los trabajos y máquinas están disponibles en el tiempo cero.

```python
# Función de aptitud(Fitness)

#Prueba 1:
#Existen 4 maquinas
#2 Trabajos
#4 operaciones por trabajo

#Individuo=[((Operación,Máquina),(Operación,Máquina),(OperaciónMáquina)),(Individuo2...),(Individuo3...)]
pruebaIndividuo1=[((1,1),(2,3),(3,4),(4,1),(5,1)),((1,1),(1,2),(2,3),(2,4),(3,1),(3,4),(4,1),(4,2),(4,3),(4,4))]

#Trabajos=[((Operación,Máquina,Tiempo),(Operación,Máquina,Tiempo),(Operación,Máquina,Tiempo)),(Trabajo2...),(Trabajo3...)
pruebaTrabajos1=[((1,1,4),(1,2,5),(2,3,7),(2,4,1),(3,1,5),(3,4,5),(4,1,1),(4,2,2),(4,3,3),(4,4,4),(5,1,4)),((1,1,1),(1,2,1),(2,3,6),(2,4,5),(3,1,1),(3,4,1),(4,1,5),(4,2,5),(4,3,2),(4,4,2))]

#ListaMaquinas ={1: 9, 2: 7, 3: 5, 4: 1}

#Penalizaciones(Castigos)
#[1]Si hay operaciones repetidas -25
#[2]Si el tiempo del trabajo es más grande que el estimado -10
#[3]Si faltan operaciones -30

def reloj(individuo,tiempoMax,trabajos,listaMaquinas):
    #print(listaMaquinas)
    for i in range(tiempoMax):
        for operacion in listaMaquinas.keys():
            if(listaMaquinas.get(operacion)!=0):
                listaMaquinas[operacion]-=1
    for operacion in listaMaquinas.keys():
        if(listaMaquinas.get(operacion)>0):
            return(False)

def fitness(individuo,tiempoMax,trabajos):
    tiempoLocal=0
    valorIndividuo=300
    for trabajo in range(len(individuo)):
        #print("Trabajo: ",trabajo+1)
        listaMaquinas={}
        listaOperaciones={}
        for operacion in range(len(individuo[trabajo])):
            #print(individuo[trabajo][operacion])
            if(listaOperaciones.get(individuo[trabajo][operacion][0])==None):# [1]
                listaOperaciones[individuo[trabajo][operacion][0]]=True
            else:
                valorIndividuo-=25
            
            if(listaMaquinas.get(individuo[trabajo][operacion][1])==None):
                for operacionPruebaTrabajo in trabajos[trabajo]:
                    if (operacionPruebaTrabajo[0]==individuo[trabajo][operacion][0] and operacionPruebaTrabajo[1]==individuo[trabajo][operacion][1]):
                        listaMaquinas[individuo[trabajo][operacion][1]]=operacionPruebaTrabajo[2]
            else:
                for operacionPruebaTrabajo in trabajos[trabajo]:
                    if (operacionPruebaTrabajo[0]==individuo[trabajo][operacion][0] and operacionPruebaTrabajo[1]==individuo[trabajo][operacion][1]):
                        listaMaquinas[individuo[trabajo][operacion][1]]+=operacionPruebaTrabajo[2]
                        
        if(reloj(individuo,tiempoMax,trabajos,listaMaquinas)==False):#[2]
            valorIndividuo-=10
        
        for i in 1,max(listaOperaciones.keys())+1:#[3]
            if(listaMaquinas.get(i)==None):
                valorIndividuo-=30
        
        #print(valorIndividuo)
    return(valorIndividuo)
                                
            
            
#fitness(pruebaIndividuo1,12,pruebaTrabajos1)
```

#### Método de ruleta
"Los padres se seleccionan de acuerdo a su fitness. Los individuos mejores (con mayor fitness) son los que tienen mayores posibilidades de ser elegidos. Intuitivamente el proceso construye una ruleta o un "tarta" en la que cada uno de las porciones representa a un individuo. La porción de tarta que le toca a cada individuo es proporcional a su fitness. Así los individuos buenos se llevarán las mayores porciones y al revés ocurrirá con los peores."[4]<br/>

tomado de [4]

```python
import random
from random import choices
#Este código fue implementado, basándose en el pseudocódigo de [5]

#Trabajos=[((Operación,Máquina,Tiempo),(Operación,Máquina,Tiempo),(Operación,Máquina,Tiempo)),(Trabajo2...),(Trabajo3...)]
#Prueba 1:
#Existen 4 maquinas
#2 Trabajos
#4 operaciones por trabajo
prueba1=[((1,1,4),(1,2,5),(2,3,7),(2,4,1),(3,1,5),(3,4,5),(4,1,1),(4,2,2),(4,3,3),(4,4,4)),((1,1,1),(1,2,1),(2,3,6),(2,4,5),(3,1,1),(3,4,1),(4,1,5),(4,2,5),(4,3,2),(4,4,2))]

#Algoritmo genético
def algoritmoGenetico(trabajos,tiempoMax,NumeroGeneraciones):
    listaOperaciones={}
    for trabajo in range(len(trabajos)):
        #print("Trabajo: ",trabajo+1)
        for operacion in range(len(trabajos[trabajo])):
            if(listaOperaciones.get(trabajos[trabajo][operacion][0])==None):
                listaOperaciones[trabajos[trabajo][operacion][0]]=True
            #print(trabajos[trabajo][operacion])
            
    #Poblacion Inicial
    poblacionInicial=10
    poblacion=[]
    for i in range(poblacionInicial):
        individuo=[]
        for trabajo in trabajos:
            trabajoIndividuo=[]
            for _ in range(max(listaOperaciones.keys())):
                crearTrabajo=trabajo[random.randint(0,len(trabajo)-1)]
                trabajoIndividuo.append((crearTrabajo[0],crearTrabajo[1]))
            individuo.append(trabajoIndividuo)
        poblacion.append(individuo)
    
    valorIndividuos=[]
    for individuo in poblacion:
        valorIndividuos.append(fitness(individuo,tiempoMax,trabajos))
    #print(poblacion)
    #print(valorIndividuos)
    terminar=False
    valorIndividuosDescendientes=[]
    nuevaGeneracion=[]
    indiceGeneracion=1
    for _ in range(NumeroGeneraciones):
        #Producir nueva generacion
        for indiceNuevaGeneracion in range(round(len(poblacion)/2)):#
            #Ciclo Reproductivo (probabilidad de seleccion proporcional a la funcion de evaluacion del individuo)
            padre=choices(poblacion, valorIndividuos)
            madre=choices(poblacion, valorIndividuos)
            #Cruzar
            descendientes=[]
            for trabajo in range(len(madre)):
                gen=0
                descendienteTrabajos1=[]
                descendienteTrabajos2=[]
                for operaciones in range(len(madre[trabajo])):
                    descendienteOperaciones1=[]
                    descendienteOperaciones2=[]
                    for operacion in range(len(madre[trabajo][operaciones])):
                        if(gen==0):
                            descendienteOperaciones1.append(madre[trabajo][operaciones][operacion])
                            descendienteOperaciones2.append(padre[trabajo][operaciones][operacion])
                            gen=1
                        else:
                            descendienteOperaciones1.append(padre[trabajo][operaciones][operacion])
                            descendienteOperaciones2.append(padre[trabajo][operaciones][operacion])
                            gen=0
                    descendienteTrabajos1.append(descendienteOperaciones1)
                    descendienteTrabajos2.append(descendienteOperaciones2)
            descendientes.append(descendienteTrabajos1)
            descendientes.append(descendienteTrabajos2)
            #Mutar
            for descendiente in range(len(descendientes)):
                for trabajo in range(len(descendientes[descendiente])):
                    for operaciones in range(len(descendientes[descendiente][trabajo])):
                        mutacion=choices([descendientes[descendiente][trabajo][operaciones],(random.randint(1,max(listaOperaciones.keys())),descendientes[descendiente][trabajo][operaciones][0])], [0.85, 0.15])
                        descendientes[descendiente][trabajo][operaciones]=mutacion[0]
            #Computar
            for individuo in descendientes:
                valorIndividuosDescendientes.append(fitness(individuo,tiempoMax,trabajos))
            nuevaGeneracion.append(descendientes[0])
            nuevaGeneracion.append(descendientes[1])
        print("Nueva Generacion",indiceGeneracion)
        indiceGeneracion+=1
        print(nuevaGeneracion)
        print(valorIndividuosDescendientes)
        poblacion=nuevaGeneracion
        valorIndividuos=valorIndividuosDescendientes
        nuevaGeneracion=[]
        valorIndividuosDescendientes=[]
        
        for tiempoSeleccion in range(len(valorIndividuos)):
            if(valorIndividuos[tiempoSeleccion]<=tiempoMax):
                print(poblacion[tiempoSeleccion])
algoritmoGenetico(prueba1,12,5)
```
