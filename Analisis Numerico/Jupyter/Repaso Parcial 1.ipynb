```octave
%biseccion
function y = biseccion(g,a,b,tol,N) %Metodo de la biseccion

fa = g(a); %Evaluacion en el primer punto
fb = g(b); %Evaluacion del segundo punto

if (fa*fb)>=0 %Verificacion del signo de los puntos
    disp('Verifique la seleccion de sus puntos')
    return
end

i = 0; %Controlador de iteraciones
while i <= N
    c = (a+b)/2; %Busqueda del nuevo punto
    fc = g(c); %Evaluacion en el nuevo punto
    if fc == 0 | abs(a-c)<tol %verificacion de si el punto
        fprintf('Un cero de la funcion es: %f\n con %d iteraciones y un error de %f\n',c,i,abs(a-c))
        break
    elseif fc*fa<0 %Asignacion del los nuevos valores
            b = c;
            fb = fc;
    else
                a = c;
                fa = fc;
    end
    if i == N %Si llega al limite de iteraciones
        fprintf ('Despues de %d iteraciones el método no funcionó\n',N)
        break
    else
            i=i+1;
    end
end
end
            
                
```

```octave
function y = falsap(g,a,b,tol,N) %Metodo de la fp

fa = g(a); %Evaluacion en el primer punto
fb = g(b); %Evaluacion del segundo punto

if (fa*fb)>=0 %Verificacion del signo de los puntos
    disp('Verifique la seleccion de sus puntos')
    return
end

i = 0; %Controlador de iteraciones
while i <= N
    c = b-fb*(b-a)/(fb-fa); %Busqueda del nuevo punto
    fc = g(c); %Evaluacion en el nuevo punto
    if fc == 0 | abs(a-c)<tol %verificacion de si el punto
        fprintf('Un cero de la funcion es: %f\n con %d iteraciones y un error de %f\n',c,i,abs(a-c))
        break
        elseif fc*fa<0 %Asignacion del los nuevos valores
            b = c;
            fb = fc;
    else
                a = c;
                fa = fc;
    end
    if i == N %Si llega al limite de iteraciones
        fprintf ('Despues de %d iteraciones el método no funcionó\n',N)
        break
    else
            i=i+1;
    end
end
end
```

```octave
function y=Puntofijo(g,X0,tol,N) %Metodo del punto fijo (Funcion, Punto inicial, Error o tolerancia, Numero máximo de iteraciones) 
Xn = g(X0); %Primera evaluacion
En = abs(Xn-X0)/abs(Xn); %Primer error.
for i=0:N %iteraciones 
    if En < tol %Comparacion de errores
        y = Xn; % devolucion del valor
        fprintf('Su punto fijo es %d con error de %d\n ',y,En)
        break
    else
        X0 = Xn; %nuevo punto
        Xn = g(Xn); %evaluacion del nuevo punto
        En = abs(Xn-X0)/abs(Xn); %error del nuevo punto
    end
    if i==N %Finalizacion si llega al limite de iteraciones
        fprintf('Despues de %d intentos no se encontro su punto fijo \n',N)
        break
    end
end
end
```
